```python
import vlc
from vlc import Instance
import os,time
class VLC:
    ismute = False                                  
    volume = 40                                         #Initial Volume
    handle = None                                       #Player handel
 
    def __init__(self):
        self.Player = Instance('--loop')
        
    def addPlaylist(self):
        self.mediaList = self.Player.media_list_new()
        path = r"C:\Users\dell5567\Desktop\engsong"     #Change the path to your music folder
        songs = os.listdir(path)
        for s in songs:
            self.mediaList.add_media(self.Player.media_new(os.path.join(path,s)))
        self.listPlayer = self.Player.media_list_player_new()
        self.listPlayer.set_media_list(self.mediaList)
    
        self.handle = self.listPlayer.get_media_player()     #Get the handel
        print(self.handle.audio_set_volume(self.volume))     #Seting the Initial Volume to media player
        
    def play(self):
        if self.ismute:
            print("Alert Volume is mute,unmute it",self.ismute)
            
        print("Playing song")
        self.listPlayer.play()
    def next(self):
        if self.ismute:
            print("Alert Volume is mute,unmute it")
        print("Playing next song")
        self.listPlayer.next()
    def pause(self):
        print("Pausing song")
        self.listPlayer.pause()
    def previous(self):
        if self.ismute:
            print("Alert Volume is mute,unmute it")
        print("Playing previous song")
        self.listPlayer.previous()
    def stop(self):
        print("Stoping the song")
        self.listPlayer.stop()
    def mute(self):
        self.ismute  = True
        print("Volume muted")
        self.handle.audio_set_volume(0)
    def unmute(self):
        print("Volume unmuted")
        #self.handle.audio_set_mute(False) Some times this function may not work so I'm seting manually
        self.handle.audio_set_volume(self.volume)
        self.ismute  = False
    def volumeup(self):
        self.volume = self.volume + 20         #You can change according to your wish (50)
        if self.volume > 100:                  #You can set volume upto 450+,Aterwards you will hear noise,
             self.volume = 100                 #So better to restrict to 450(It Depends on you are laptop Output) 
        if self.ismute:
            print("Alert Volume is mute,unmute it")
        else:
            self.handle.audio_set_volume(self.volume),
            print('Volume up',self.volume)
    def volumedown(self):
        self.volume = self.volume - 20
        if self.volume < 0:
            self.volume = 0
        if self.ismute:
            print("Alert Volume is mute,unmute it")
        else:
            self.handle.audio_set_volume(self.volume),
            print('Volume down',self.volume)
    def name(self):
        String = self.handle.get_media().get_mrl()
        String = String.split('/')[-1]
        print("Name of the song : ",String)
```

```python
player = VLC()
player.addPlaylist()

player.play()
time.sleep(5)

player.next()
time.sleep(5)

player.pause()
time.sleep(5)

player.previous()
time.sleep(5)


player.mute()
time.sleep(5)


player.volumeup()
time.sleep(5)

player.unmute()
time.sleep(5)

player.volumeup()
time.sleep(5)

player.volumedown()
time.sleep(5)

player.volumeup()
time.sleep(5)

player.next()
time.sleep(5)

player.volumeup()
time.sleep(5)


player.volumeup()
time.sleep(5)

player.name()
time.sleep(5)

player.stop()
```
